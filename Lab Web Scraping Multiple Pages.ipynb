```python
from bs4 import BeautifulSoup   ### pip install bs4
import pandas as pd
import requests
```

```python
url = 'https://es.wikipedia.org/wiki/Anexo:%C3%93scar_a_la_mejor_actriz'
```

```python
headers = {"User-agent": 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/79.0.3945.130 Safari/537.36'}
```

```python
respuesta = requests.get(url, headers=headers)
```

```python
respuesta
```

```python
all_tables = pd.read_html(respuesta.content, encoding = 'utf8')
```

```python
print(f'Total de tablas encontradas: {len(all_tables)}')
```

```python
for i, table in enumerate(all_tables):
    print(f"Tabla {i + 1}:")
    print(table)

```

```python
matched_table = pd.read_html(respuesta.text, match='Katharine Hepburn')
```

```python
print(f'Total de tablas encontradas: {len(matched_table)}')
```

```python
Katharine_Hepburn = matched_table[0]
```
