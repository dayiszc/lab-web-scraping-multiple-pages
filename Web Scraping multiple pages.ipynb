```python
artists_2020 = []
songs_2020 = []
id = 0

for i in range(1, 6):
    if i == 1:
        id = 1054377950
    elif i == 2:
        id = 1054378275
    elif i == 3:
        id = 1054379062
    elif i == 4:
        id = 1054379661
    else:
        id = 1054380365
    r = requests.get(f"https://www.npr.org/2020/12/03/934634998/the-100-best-songs-of-2020-page-{i}", headers=header)
    soup = BeautifulSoup(r.content, 'html.parser')
    artists_2020.extend([artist.get_text(strip=True) for artist in soup.find_all("h3", attrs={'class': 'edTag'}) if '"' not in artist.get_text() and 'Next >' not in artist.get_text() and '< Previous' not in artist.get_text() and '< Previous•Next >' not in artist.get_text()])
    songs_2020.extend([song.get_text(strip=True).replace('"', '') for song in soup.find_all("h3", attrs={'class': 'edTag'}) if '"' in song.get_text()])
```

```python
artists_2021 = []
songs_2021 = []
id = 0

for i in range(1, 6):
    if i == 1:
        id = 1135802083
    elif i == 2:
        id = 1135802978
    elif i == 3:
        id = 1135803422
    elif i == 4:
        id = 1135803900
    else:
        id = 1135804266
    r = requests.get(f"https://www.npr.org/2022/12/15/{id}/100-best-songs-2022-page-{i}", headers=header)
    soup = BeautifulSoup(r.content, 'html.parser')
    artists_2021.extend([artist.get_text(strip=True) for artist in soup.find_all("h3", attrs={'class': 'edTag'}) if '"' not in artist.get_text() and 'Next >' not in artist.get_text() and '< Previous' not in artist.get_text() and '< Previous•Next >' not in artist.get_text()])
    songs_2021.extend([song.get_text(strip=True).replace('"', '') for song in soup.find_all("h3", attrs={'class': 'edTag'}) if '"' in song.get_text()])
```
